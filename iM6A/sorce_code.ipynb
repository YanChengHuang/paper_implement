```python
for key,value in Dict.items():
    motif = key
    List1 = Dict[motif][-1]
    List2 = Dict[motif][0]
    List3 = Dict[motif][1]
    Score = [-1]*len(List1) + [0]*len(List2) + [1]*len(List3) #linear regression的input X (維度為此5-mer 有參與到幾次預測不管是-1(break),0(other),1(create))
    Value = List1 + List2 + List3 #linear regression的input Y
    df = pd.DataFrame({"Score":Score, "Value":Value})
    # 1024個5-mer motif存成csv
    df.to_csv("./Figure2/LongInternalExon/{}.csv".format(motif), index=0)
```

```python
Name = []
Intercept = []
Slope = []
R2 = []
Pvalue = []

for f in file:
    df = pd.read_csv(f)    
    name = f[17:22] #5-mer名稱
    x = np.asarray(df["Score"], dtype=float)
    y = np.asarray(df["Value"], dtype=float)
    x = sm.add_constant(x) #加入常數項1
    model = sm.OLS(y, x).fit()
    
    r2 = model.rsquared    
    intercept = model.params[0]
    slope = model.params[1]    
    pvalue = model.pvalues[0]
    
    Name.append(name)
    Slope.append(slope)
    Intercept.append(intercept)
    R2.append(r2)
    Pvalue.append(pvalue)
```
